### Setup: Installing and Importing Required Libraries

```python
import subprocess
import sys

# List of required packages
required_packages = [
    "pickle", "pandas", "sklearn", "numpy"
]

def install_package(package):
    """Installs a package using pip if it's not already installed."""
    try:
        __import__(package)
        print(f"{package} is already installed.")
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Check and install missing packages
for package in required_packages:
    install_package(package)
```

```python
import pickle
import pandas as pd
import numpy as np
```

### Load the Datasets (Already preprocessed)

```python
# File path
file_path = "../Dataset/sw_reviews.pkl"

with open(file_path, 'rb') as file:
  sw_reviews_df = pickle.load(file)

sw_reviews_df.head()
```

```python
# File path
file_path = "../Dataset/others_reviews.pkl"

with open(file_path, 'rb') as file:
  others_reviews_df = pickle.load(file)

others_reviews_df.head()
```

```python
complete_df = pd.concat([sw_reviews_df, others_reviews_df], ignore_index=True)
complete_df.head()
```

## TF-IDF analysis

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Import ENGLISH_STOP_WORDS from sklearn for stopword removal
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the processed review text
tfidf_matrix = vectorizer.fit_transform(complete_df['Review_Text'])

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Compute the global TF-IDF score for each word
global_tfidf_scores = np.array(tfidf_matrix.sum(axis=0)).flatten().astype('float32')

# Create a DataFrame mapping words to their global TF-IDF scores
tfidf_scores_df = pd.DataFrame({'word': feature_names, 'tfidf_score': global_tfidf_scores})

# Remove English stop words from the DataFrame
tfidf_scores_df = tfidf_scores_df[~tfidf_scores_df['word'].isin(ENGLISH_STOP_WORDS)]

# Sort in descending order
tfidf_scores_df = tfidf_scores_df.sort_values(by='tfidf_score', ascending=False)

# Print the top N words with highest TF-IDF scores
print(tfidf_scores_df.head(20))
```

## IDF Analysis

```python
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
import pickle

# Create the TF-IDF vectorizer
vectorizer = TfidfVectorizer(use_idf=True, smooth_idf=True)

# Fit and transform the text
tfidf_matrix = vectorizer.fit_transform(complete_df['Review_Text'])

# Retrieve IDF values and corresponding feature names
idf_values = vectorizer.idf_
feature_names = vectorizer.get_feature_names_out()

# Create a dictionary mapping words to their IDF values
idf_dict = dict(zip(feature_names, idf_values))

# OPTIONAL: Remove stopwords
idf_dict = {word: idf for word, idf in idf_dict.items() if word not in ENGLISH_STOP_WORDS}

# Sort the dictionary by IDF values in ascending order
idf_sorted = sorted(idf_dict.items(), key=lambda x: x[1])

# Save to pickle
with open('idf_sorted.pkl', 'wb') as file:
    pickle.dump(idf_sorted, file)

# Print sorted IDF values
for word, idf in idf_sorted:
    print(f"{word}: {idf}")
```
